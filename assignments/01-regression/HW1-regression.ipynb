```python
X = data.loc[:, data.columns != 'Price'].to_numpy()  # convert dataframe to numpy array
y = data.loc[:, data.columns == 'Price'].to_numpy()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler()
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)

y_scaler = MinMaxScaler()
y_train = y_scaler.fit_transform(y_train)
```

```python
X_train
```

```python
from sklearn import linear_model

# reg = linear_model.Lasso(alpha=1)
reg = linear_model.Ridge(alpha=0.02)
reg.fit(X_train, y_train)
```

```python
y_hat = reg.predict(X_test)
y_hat = y_scaler.inverse_transform(y_hat.reshape([-1, 1]))
print(sum(abs(y_hat - y_test))/len(y_hat))
```
